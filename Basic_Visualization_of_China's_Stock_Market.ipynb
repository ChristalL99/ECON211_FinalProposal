```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("https://raw.githubusercontent.com/ChristalL99/ECON211_FinalProposal/main/China_Stock_Data_2015.csv", index_col=0)
data
```

```
# select the data of Pingan Bank
df1 = data[data["code"]=="000001.XSHE"]
df1
```

```
# visualization of Pingan Bank in terms of Daily Money Traded and Open Price
fig = plt.figure()

x = df1["time"]
y = df1["money"]
z = df1["open"]

ax1 = fig.add_subplot(111)
lns1 = ax1.plot(x, y, label="Money Traded Daily")
ax1.set_xticks([""])
ax1.set_ylabel("Money",fontsize=11)
ax1.set_xlabel("Date", fontsize=11)
ax1.set_title("Trading Information of Pingan Bank in 2015",fontsize=12)


ax2 = ax1.twinx()
lns2 = ax2.plot(x, z, 'r', label="Daily Open Price")
ax2.set_xticks([""])
ax2.set_ylim([5,25])
ax2.set_ylabel("Open Price")
ax2.legend(loc=0)

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax2.legend(lns, labs, loc=0)


plt.show()
```
